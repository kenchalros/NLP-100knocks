```python
# タブ1文字につきスペース1文字に置換せよ．
# 確認にはsedコマンド，trコマンド，もしくはexpandコマンドを用いよ．
```

```python
with open('hightemp.txt') as f:
    # strip()で'\n'を削除，replace()でタブをスペースに変換
    content = [ s.strip().replace('\t', ' ') for s in f.readlines()]
for s in content:
    print(s)
```

```python
"""
コマンド解説
    - tr (TRanslate or TRansliterate)
        - `tr <置換対象の文字> <置換後の文字>`で置換する．
        - `tr <a> <b> <file-name>` でfile-nameに保存する．
    - その他文字列置換に使えるコマンド
        - sed (Stream EDitor)
        - expand　
"""
!grep '' hightemp.txt | tr '\t' ' '
```
